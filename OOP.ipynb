```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def startMoving(self):
        return ('I am moving forward')
    
    def stopMoving(self):
        print('I have stopped moving')
        
        
        
car = Car('Toyota', 'Camry', 2008)

print(car.make)
print(car.model)
print(car.year)
```

```python
cars = Car('bigman', 'trojan', 1995)

print(cars.make)
print(cars.model)
print(cars.year)
```

```python
cars.year = 2020
print(car.year)
print(cars.year)
```
