```python
#Ex.72

#Write a program that displays the answers for the first 100 numbers in the Fizz-Buzz game. Each Answer should be displayed on its own line. 

#1. I need to know what the first 100 numbers will be.
#2. They need to either be divisable by 5 and/or 3 
#3. I need to reject anything that isnt. 
#4. I need to do this until I have 100 of them 

'''
Select a starting number of 0
add 1 to the number and check if it is divisible by 3 and then 5.
If it is then I add it to my special storage of a list.
While I am adding the numbers to the list i need to check the length of the list is less than 100 and then stop
'''

num = 1
store = []
while len(store)<100:
    if (num%3==0) and (num%5==0):
        x = 'FizzBuzz'
    elif (num%5==0):
        x = 'Buzz'
    elif (num%3==0):
        x = 'Fizz'
    else:
        x = num
    
    store.append(x)
    num += 1

for i in store:
    print(i)    
```

```python
#Ex 73
#Implement a Caeser cipher
#Allow the user to supply the shift amount
#ensure upper and lower case encapsulation
#It should encorporate negative shift values
#It should display the shifted message

'''
have a list of letters
find the letter in the upper or lower list
tell the program to add the new letter to a new list 
combine the new list togther to mke a string 
'''

upLetters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
loLetters = 'abcdefghijklmnopqrstuvwxyz'

string =''
place =[]
word = input('What would you like your message to say? ')
shift = int(input('What would you like your shift pattern to be? I can be both positive and negative.'))

while shift > 56 or shift <-56:
    shift = int(input('Please choose a nuber between -26 and 26. '))
    
        
for letta in word:
    if letta in upLetters:
        y = upLetters.find(letta)
        y = y + shift 
        if y > 25:
            y = (y - 26)
        elif y < 0:
            y = (26+y)
        newletta = upLetters[y]
        place.append(newletta)
    elif letta in loLetters:
        y = loLetters.find(letta)
        y = y + shift 
        if y > 25:
            y = (y - 26)
        elif y < 0:
            y = (26+y)
        newletta = loLetters[y]
        place.append(newletta)
    else:
        place.append(letta)
string=string.join(place)
print('')
print('Your new message is: ', string)
```
